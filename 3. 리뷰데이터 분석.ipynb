# 3. 리뷰 데이터 분석

## 1) 라이브러리&데이터 불러오기

```python
import pandas as pd
from konlpy.tag import Okt
import nltk
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
all_token_txt = open('Datasets/06_all_token_data.txt', 'r', encoding='utf-8')

token_ko = []

for line in all_token_txt.readlines():
    token_ko.append(line[:-1])

all_token_txt.close()
```

```python
all_token = nltk.Text(token_ko)
```

## 2) 빈도수 상위 20개 단어

```python
plt.figure(figsize=(15,10))
plt.title('경기도 관광지 리뷰데이터 상위 20개 단어')
all_token.plot(20)
```

## 3) 워드클라우드

```python
fd = all_token.vocab()
```

```python
wc = WordCloud(font_path='c:/Windows/Fonts/NanumGothic.ttf',
               width=1000,
               height=1000,
               max_words=100,
               max_font_size=300,
               background_color='white',
               colormap='Dark2',
              ).generate_from_frequencies(fd)
plt.figure(figsize=(15,10))
plt.imshow(wc)
plt.axis('off')
plt.show()
```
