# 4. 리뷰데이터 분석(2)

## 1) 라이브러리&데이터 불러오기

```python
import pandas as pd
from konlpy.tag import Okt; okt = Okt()
import nltk
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import numpy as np
```

```python
df_token = pd.read_csv('Datasets/02_Preprocessing.csv', encoding='UTF-8', index_col=0)
```

```python
df_token
```

## 2) 리뷰 길이에 따른 정렬

#### 2-1. 리뷰 길이 컬럼 만들기

```python
length = []

for i in df_token['내용']:
    length.append(len(i))
    
df_token['길이'] = length
df_token.head()
```

#### 2-2. 리뷰 길이에 따른 내림차순 정렬

```python
df_sort = df_token.sort_values(by=['길이'], axis=0, ascending=False)
df_sort.head()
```

- 인덱스 재정렬

```python
df_sort.reset_index(drop=True, inplace=True)
```

- 상위 10개

```python
df_sort.head(10)
```

- 하위 10개

```python
df_sort.tail(10)
```

## 3) 토큰화 함수 만들기

#### 3-1. 불용어 txt 불러오기

```python
sw_txt = open('Datasets/05_StopWords(3차).txt', 'r', encoding='utf-8')
sw_list = []

for line in sw_txt.readlines():
    sw_list.append(line[:-1])

sw_txt.close()
```

#### 3-2. 토큰화 + 불용어 처리 + 상위 10개 단어 그래프 + 워드클라우드 함수

```python
def make_token(name):
    text = df_sort[df_sort['수집채널명'] == name]['내용'].values[0]
    # 토큰화
    token_text = okt.morphs(text)
    token = nltk.Text(token_text)
    print('수집채널명:', name)
    print('전체 토큰 수:', len(token.tokens))
    
    # 불용어 처리
    token = [each_word for each_word in token if each_word not in sw_list]
    token = nltk.Text(token)
    
    fd = token.vocab()
    
    # 그래프 그리기
    plt.figure(figsize=(10,8))
    plt.title(f'{name} 관광지 리뷰데이터 상위 10개 단어')
    token.plot(10)
    
    # 워드클라우드
    plt.figure(figsize=(15,10))
    wc = WordCloud(font_path='c:/Windows/Fonts/NanumGothic.ttf',
               width=1000,
               height=1000,
               max_words=100,
               max_font_size=300,
               background_color='white',
               colormap='Dark2',
              ).generate_from_frequencies(fd)
    plt.axis('off')
    plt.imshow(wc)
```

## 4) 항목별 리뷰데이터 분석(상위 빈도수, 워드클라우드)

#### 4-1. 상위 3개

```python
make_token('에버랜드')
```

```python
make_token('DMZ')
```

```python
make_token('한국민속촌')
```
