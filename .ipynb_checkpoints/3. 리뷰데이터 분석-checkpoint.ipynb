# 3. 리뷰 데이터 분석

## 1) 라이브러리&데이터 불러오기

```python
import pandas as pd
from konlpy.tag import Okt
import nltk
import matplotlib.pyplot as plt
from wordcloud import Wordcloud
```

```python
all_token_txt = open('Datasets/all_token_data.txt', 'r', encoding='utf-8')

token_ko = []

for line in all_token_txt.readlines():
    token_ko.append(line[:-1])

all_token_txt.close()
```

```python
all_token = nltk.Text(token_ko)
```

## 2) 빈도수 상위 20개 단어

```python
plt.figure(figsize=(15,10))
plt.title('경기도 관광지 리뷰데이터 상위 20개 단어')
all_token.plot(20)
```

## 3) 워드클라우드

```python
wc = WordCloud(width=1000, height=600, background_color="white", random_state=0)

plt.imshow(wc.generate_from_frequencies(all_token))
plt.axis("off")
plt.show()
```
