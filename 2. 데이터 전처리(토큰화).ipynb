```python
df_token = pd.read_csv('Datasets/02_Preprocessing.csv', encoding='UTF-8', index_col=0)
```

## 3) 불용어 처리

- 전체 리뷰데이터 합치기

```python
content_text = ''

for each_text in df_token['내용']:
    content_text = content_text + each_text + '\h'
```

- 전체 리뷰데이터 토큰화

```python
token_ko = okt.morphs(content_text)

all_token = nltk.Text(token_ko)
print('전체 토큰의 수:', len(all_token.tokens))
print('중복 제외한 전체 토큰의 수:', len(set(all_token.tokens)))
```

#### 3-1. 1차 불용어 처리
- 빈출 상위 100개 단어 조회

```python
all_token.vocab().most_common(100)
```

- 불용어 txt 불러오기(상위 100개에서 필요없는 단어로 작성)

```python
sw_txt = open('Datasets/03_StopWords(1차).txt', 'r', encoding='utf-8')

sw_list = []

for line in sw_txt.readlines():
    sw_list.append(line[:-1])

sw_txt.close()
```

```python
print('불용어 처리할 단어 수:', len(sw_list))
```

- 불용어 제거하기

```python
token_ko = [each_word for each_word in token_ko if each_word not in sw_list]
```

- 불용어 제거 후 빈출 상위 100개 단어 조회

```python
all_token = nltk.Text(token_ko)
all_token.vocab().most_common(100)
```
